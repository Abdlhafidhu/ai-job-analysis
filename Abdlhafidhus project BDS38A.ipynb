```python
#5.Describe the dataset extensively
print("\nExtensive description of the dataset:")
print(df.info())
print("\nSummary statistics of the dataset:")
print(df.describe(include='all'))
```

```python
#6.Display the number of null values in the dataset and the total count of it
null_values = df.isnull().sum()
print("Number of null values in each column:")
print(null_values)
```

```python
total_null_values = null_values.sum()
print("\nTotal count of null values in the dataset:")
print(total_null_values)
```

```python
#7.Identify the columns having null values and remove them
```

```python
columns_with_nulls = null_values[null_values > 0].index
print("\nColumns with null values:")
print(columns_with_nulls)
df_cleaned = df.drop(columns=columns_with_nulls)
```

```python
# Display the cleaned dataset
print("\nCleaned dataset (top 5 records):")
print(df_cleaned.head())
print("\nDimensionality of the cleaned dataset (rows, columns):")
print(df_cleaned.shape)
```

```python
# 8.check the size of dataset after cleaning and compare with the size before cleaning
print("\nSize of the dataset before cleaning (rows, columns):")
print(df.shape)
print("Size of the dataset after cleaning (rows, columns):")
print(df_cleaned.shape)
```

```python
#9.Find out the count of male and female penguins
sex_counts = df['sex'].value_counts()
print("\nCount of male and female penguins:")
print(sex_counts)
```

```python
#10.Find out the count of species 
species_counts = df['species'].value_counts()
print("\nCount of each species:")
print(species_counts)
```

```python
#11. Use a countplot graph to display the number of male and female penguins
#Give the plot a title "Count of male and female penguins"
#Give a xlabel and ylabel 
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, hue='sex', palette='Set2', legend=False)
plt.title("Count of Male and Female Penguins", fontsize=16)
plt.xlabel("Sex", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```

```python
#12. For all the penguins display the bill_length_mm and bill_depth_mm using pairplot
sns.pairplot(df, vars=['bill_length_mm', 'bill_depth_mm'], hue='species', palette='Set2')
plt.suptitle("Pair Plot of Bill Length and Bill Depth by Species", y=1.02, fontsize=16)
plt.show()
```

```python
#13.Create a visualization to display the flipper_length_mm and body_mass_g for the penguin dataset.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=df, palette='Set2')
plt.title("Scatter Plot of Flipper Length vs Body Mass by Species", fontsize=16)
plt.xlabel("Flipper Length (mm)", fontsize=14)
plt.ylabel("Body Mass (g)", fontsize=14)
plt.show()
```

```python
#14. Create a histogram with the flipper_length_mm and bins = 10
plt.figure(figsize=(10, 6))
sns.histplot(df['flipper_length_mm'], bins=10, kde=True, color='skyblue')
plt.title("Histogram of Flipper Length (mm)", fontsize=16)
plt.xlabel("Flipper Length (mm)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

```python
#15. Create a joint point with body_length_mm, body_depth_mm for penguins_cleaned dataset
# Drop rows with null values to create a cleaned dataset
df_cleaned = df.dropna()
# Create a joint plot for 'bill_length_mm' and 'bill_depth_mm'
sns.jointplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=df_cleaned,
    kind='scatter',  # You can change this to 'reg' for a regression line or 'kde' for a density plot
    height=6,
    color='skyblue')

# Add a title to the plot
plt.suptitle("Joint Plot of Bill Length and Bill Depth", y=1.02, fontsize=16)

# Show the plot
plt.show()
```

```python
# Create a pairwise plot for all numerical values
sns.pairplot(df_cleaned, hue='species', palette='Set2')

# Add a title to the plot
plt.suptitle("Pairwise Plot of All Numerical Values in the Penguins Dataset", y=1.02, fontsize=16)

# Show the plot
plt.show()
```

```python
#17.Check if there exists any outliers based on the boxplot based on numerical values  of penguins_cleaned dataset
# Create boxplots for all numerical columns
numerical_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Plot boxplots
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)  # Adjust the layout based on the number of numerical columns
    sns.boxplot(x=df_cleaned[column])
    plt.title(f'Boxplot of {column}')

# Add a main title
plt.suptitle("Boxplots of Numerical Values in the Penguins Dataset", y=1.02, fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```

```python
#18. Create a violin plot to display the body_mass_g of different species based on sex
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='body_mass_g', hue='sex', data=df_cleaned, palette='Set2', split=True)

# Add a title and labels
plt.title("Violin Plot of Body Mass by Species and Sex", fontsize=16)
plt.xlabel("Species", fontsize=14)
plt.ylabel("Body Mass (g)", fontsize=14)

# Show the plot
plt.show()
```
